```python
billboard.rename(columns = {
    'artist.inverted' : 'artist_name',
    'track' : 'track_name',
    'time' : 'track_length',
    'date.entered' : 'date_entered',
    'date.peaked' : 'date_peaked'},
    inplace = True)
```

```python
billboard.columns
```

```python
for col in billboard.columns:
    if any([character in col for character in ['$','&','%','#','.','x']]):
        newcol = ''.join([character for character in col if character not in ['$','&','%','#','.','x']])
        billboard.rename(columns={col:newcol}, inplace=True)

billboard.columns
```

```python
#Changing the 'date_entered' and 'date_peaked' to DateTime format.
billboard.date_entered = pd.to_datetime(billboard.date_entered)
billboard.date_peaked = pd.to_datetime(billboard.date_peaked)
```

```python
billboard = billboard.dropna(axis = 1, how = 'all')
print billboard.head(5)
```
